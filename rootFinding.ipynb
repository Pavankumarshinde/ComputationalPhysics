**Newton Rapsion**


```python
import numpy as np
from math import tan, exp, cos
import matplotlib.pyplot as plt

def f(x):
    return tan(x) - exp(x)

def df(x):
    return (1/np.cos(x))**2 - exp(x)

x = np.linspace(-10, 10, 100)
y = np.vectorize(f)(x) 
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = tan(x) - exp(x)')
plt.axhline(0, color='gray', linestyle='--')
plt.ylim(-10, 10) 

def newtonRapson(x0,f,tol):
    err=abs(f(x0))
    while(err > tol):
        x1= x0 -f(x0) /df(x0)
        err=f(x1)
        x0=x1
    return x0
x0=newtonRapson(1.55,f,10e-5)
plt.scatter(x0,f(x0))
```

**Numerical Integration**

```python
from math import sin, tan

def f(x):
    if(x!=0):
        return sin(x)/x
    else:
        return 1

a=0
b=1

# trapezoid
def trapezoidIntegration(a,b,f):
    N=1000000
    delta_x=(b-a)/N
    sum=f(a)/2
    for i in range(1 , N):
        sum+=f(a + delta_x* i)
    sum+=f(N)/2
    sum=sum*delta_x

    
    return sum
def riemannIntegration(a,b,f):
    sum=0
    N=1000000
    delta_x=(b-a)/N
    for i in range(0,N):
        sum+=f(a+delta_x*i)
    sum= sum * delta_x
    return sum
print(trapezoidIntegration(a,b,f))

print(riemannIntegration(a,b,f))
    
```
